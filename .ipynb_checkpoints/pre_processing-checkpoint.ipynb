```python
import pandas as pd
import numpy as np
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('results.csv',header=0,names=['BRAND','CAR','FUEL','BODY_TYPE','INFO1','INFO2','PRICE1','PRICE2','PRICE3'],error_bad_lines=False)
```

```python
df.head()
```

```python
len(df)
```

```python
def pre_processing(df):
    df['Model']=df['CAR'].apply(lambda x:x.split(' ')[2])
    df['Year_Manufacture']=df['CAR'].apply(lambda x: re.findall(r'\d{4}',x)[0])
    df['Clean']=df['INFO2'].apply(lambda x:re.sub(r'\d{3}','',x))
    df['Owners']=df['Clean'].apply(lambda x: re.findall(r'\d',x)[0])
    df['Km_Driven']=df['INFO1'].apply(lambda x: re.findall(r'\d+',x)[0]+'000')
    df['pRICE1']=df['PRICE1'].apply(lambda x: re.findall(r'\d+',x)[0])
    df['pRICE2']=df['PRICE2'].apply(lambda x: re.findall(r'\d+',x)[0])
    df['Price']=df['pRICE1']+df['pRICE2']+'000'
    df=df[['BRAND','Model','Year_Manufacture','FUEL','BODY_TYPE','Owners','Km_Driven','Price']]
    return df
```

```python
df_clean=pre_processing(df)
```
